* Crear una clase Usuario
* lista de objetos de la clase Usuario
* Crear un menú de opciones donde las opciones sean:

  * Opción 1: Imprimir todos los usuarios de la lista

  * Opción 2: Imprimir todos los usuarios ordenados por edad descendente. Opcional: pedir por input si quiere ascendente o descendente

  * Opción 3: Imprimir un usuario por su email: Ejemplo, con input preguntamos qué usuario quiere ver, y ese cuyo id coincida lo imprimimos. (opcional, quien quiera puede pensarlo con un número int que se genere en base al máximo sumando uno o con una secuencia)

  * Opción 4: Crear un nuevo objeto usuario, se le puede pedir los datos con input(): 
    * nombre (str)
    * email (str)
    * edad (int)
    * altura (float)
    * estudiante (bool)
    * Opcional: fecha cumpleaños con date (Leer pedir primero dia, luego mes y luego año)

  * Opción 5: Actualizar un usuario que ya exista

  * Opción 6: borrar un usuario por su email

  * Opción 7: borrar todos los usuarios

  * Opción 8: Salir


- Extra opcional: Dividir en funciones cada operación para separarlas del menú
- Extra opcional: gestión de errores con try-except y bucles para reintentos
- Extra opcional: guardar los usuarios en un csv según se crean en la lista.
- Extra opcional: el usuario tenga un atributo que sea un objeto de otra clase, una clase Address (street, city, country).


- Ver usuarios / usuario: 25 % 
- Crear: 25 %

```python
# Importaciones
from datetime import datetime # para trabajar con fechas
import random # Para crear numeros aleatorios

# Definimos la clase y sus atributos de objeto. 
# Los datos del usuario se deben pasar como parametros al crearlo
class Usuarios:
    def __init__(self, nombre, apellido, dni, altura, email, estudiante): # Metodo que llamamos cuando creamos un objeto usuario
        self.nombre = nombre
        self.apellido = apellido
        self.dni = dni
        self.altura = float(altura) # Convertimos la altura en float
        self.edad = None  # La edad se inicializa en "None" porque se calcula despues
        self.email = email
        self.estudiante = estudiante
        self.id = random.randint(0, 999)  # Se genera un "ID" único

    # Método para calcular la edad
    def calcular_edad(self):
        nacimiento = input(f"Dime tu año de nacimiento de {self.nombre}: ") # Pedimos el año de nacimiento
        now = datetime.now() # Sacamos la fecha actual
        self.edad = now.year - int(nacimiento) # Calculamos los años extrayendo el año actual de la fecha y le restamos el año de nacimiento que nos proporciona el usuario y la establecemos en el parametro

    # Para mostrar datos del usuario por pantalla
    # Si no hay edad muestra "desconocida"
    # Convertimos el parametro de estudiante a "si" o "no"
    def __str__(self):
        return (f"Usuarios:\n Nombre = {self.nombre}\n Apellido = {self.apellido}\n"
                f"DNI = {self.dni}\n Altura = {self.altura}m\n Edad = {self.edad if self.edad else 'Desconocida'} años\n"
                f"Email = {self.email}\n Estudiante = {'Sí' if self.estudiante else 'No'}\n ID = {self.id}\n")


# Obtenemos el año actual para calcular la edad de los usuarios de ejemplo
hoy = datetime.now().year

# Usuarios de ejemplo ya añadidos
usuario1 = Usuarios(nombre="Pepe", apellido="Martinez", dni="8907654Y", altura=1.75, email="usuario1@gmail.com", estudiante=True)
usuario1.edad = hoy - 2012

usuario2 = Usuarios(nombre="Luis", apellido="Gonzalez", dni="6923837Q", altura=1.90, email="usuario2@gmail.com", estudiante=False)
usuario2.edad = hoy - 1979

usuario3 = Usuarios(nombre="Juan", apellido="Lopez", dni="2840281P", altura=1.68, email="usuario3@gmail.com", estudiante=True)
usuario3.edad = hoy - 2000

#Lista de usuarios
usuarios = [usuario1, usuario2, usuario3]

# Menú
menu = """
Bienvenido, elije una de las opciones:\n
    1.- Ver todos los usuarios
    2.- Ordenar usuarios por edad descendente
    3.- Buscar usuario por Email
    4.- Crear nuevo usuario
    5.- Actualizar usuario existente
    6.- Borrar usuario
    7.- Borrar todos los usuarios
    8.- Salir
"""

# Iniciamos un bucle infinito para mostrar el menu. No se para de mostrar el menu hasta que salimos con la opcion 8
while True:
    opcion = int(input(menu))
    if 1 <= opcion <= 8: # Confirmamos que se elija una opcion de la 1 a la 8
        match opcion:
            
            case 1:
                # Mostrar todos los usuarios
                # Recorremos la lista para comprobar si esta vacia, si no lo está imprime todos los datods de los usuarios con el metodo __str__()
                if usuarios:
                    for usuario in usuarios:
                        print(usuario)
                else:
                    print("No hay usuarios registrados.")
                    
            case 2:
                # Ordenar usuarios por edad descendente
                # Usamos la funcion "sorted()" para ordenar la lista
                usu_orden_descen = sorted(usuarios, key=lambda o: o.edad if o.edad is not None else -1, reverse=True)
                # Aqui ordenamos por la edad, si no tiene edad se le pone -1 para ponerlo al final. Con "reverse=True" ordenamos de mayor a menor
                
                for usuario in usu_orden_descen:
                    print(usuario)
                # Recorremos la lista y mostramos los usuarios
                
            case 3:
                # Buscar usuario por Email
                correo = input("Correo del usuario a buscar: ")
                usuario_encontrado = next((u for u in usuarios if u.email.lower() == correo.lower()), None)
                # Buscamos al usuario con el email introducido con el metodo "next() que recorre la lista para obtener sus elementos". Pasamos los correos a minuscula
                if usuario_encontrado:
                    print(usuario_encontrado)
                else:
                    print("Usuario no encontrado.")
                    
            case 4:
                # Crear un nuevo usuario
                # Solicitamos todos los datos para crear un nuevo usuario
                nombre = input("Introduce el nombre de usuario: ")
                apellido = input("Introduce el apellido de usuario: ")
                altura = float(input("Introduce la altura del usuario (en metros): ")) # Convertimos la altura a float
                dni = input("Introduce el DNI de usuario: ")
                email = input("Introduce el Email de usuario: ")
                estudiante = input("¿Es estudiante? (sí/no): ").strip().lower() == "sí"
                # Convertimos si es estudiante a booleano y compara con un "si" para poner True o False.
                # Con "strip()" quitamos los espacion en blanco y con "lower()" lo convertimos en minuscula
                
                nuevo_usuario = Usuarios(nombre, apellido, dni, altura, email, estudiante) # Creamos el objeto
                nuevo_usuario.calcular_edad()  # Calcula la edad del usuario
                usuarios.append(nuevo_usuario) # Añadimos el nuevo usuario
                print("El usuario se ha creado satisfactoriamente.")# Se muestra mensaje de confirmacion
                
            case 5:
                # Actualizar usuario existente
                correo = input("Introduce el email del usuario a actualizar: ") # Buscamos al usuario
                usuario_encontrado = next((u for u in usuarios if u.email.lower() == correo.lower()), None) # Recorremos la lista
                if usuario_encontrado: # Si lo encuentra solicita los datos por pantalla
                    usuario_encontrado.nombre = input("Nuevo Nombre: ")
                    usuario_encontrado.apellido = input("Nuevo Apellido: ")
                    usuario_encontrado.dni = input("Nuevo DNI: ")
                    usuario_encontrado.altura = float(input("Nueva altura (en metros): "))
                    usuario_encontrado.email = input("Nuevo Email: ")
                    usuario_encontrado.estudiante = input("¿Es estudiante? (sí/no): ").strip().lower() == "sí"
                    print("Usuario actualizado.")
                else:
                    print("Usuario no encontrado.")
                    
            case 6:
                # Borrar un usuario
                correo = input("Introduce el email del usuario a borrar: ")
                usuarios = [u for u in usuarios if u.email.lower() != correo.lower()] 
                # Hacemos una compresion de lista
                # [expresion for elemento in iterable if condicion]
                # expresion: Es lo que se añade a la nueva lista.
	            # for elemento in iterable: Itera sobre cada elemento del iterable.
		        # if condicion (opcional): Filtra los elementos que cumplen la condición.

		        # u: Es cada elemento en la lista original usuarios.
		        # for u in usuarios: Itera sobre cada usuario en la lista usuarios.
		        # if u.email.lower() != correo.lower(): Incluye en la nueva lista solo los usuarios cuyo email no coincida el correo.
                
                print("Usuario borrado.")
                
            case 7:
                # Borrar todos los usuarios
                usuarios.clear()
                print("Todos los usuarios han sido borrados.")
                
            case 8:
                # Salir del programa
                print("¡Adiós!")
                break # Salimos del bucle while true
    else:
        print("No has elegido una opción correcta.")
```
